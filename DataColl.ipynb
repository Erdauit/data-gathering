```python
import numpy as np
```

```python
#1 Создание массива
arr = np.array([1, 2, 3, 4, 5])
```

```python
#2 Создание матрицы
A = np.array([[1, 2], [3, 4]])
print("Матрица A:\n", A)
```

```python
#3 Транспорзиция матрицы
A_transpose = A.T
print("Транспозиция A:\n", A_transpose)
```

```python
#4 Сумма матриц
B = np.array([[5, 6], [7, 8]])
C = A + B
print("Сумма матриц A и B:\n", C)
```

```python
#5 Разность матриц
D = B - A
print("Разность матриц B и A:\n", D)
```

```python
#6 Умножение матриц
E = A * B
print("Поэлементное умножение матриц A и B:\n", E)
```

```python
#7 Усножение матрицы на скаляр
scalar = 2
F = A * scalar
print("Матрица A, умноженная на 2:\n", F)
```

```python
#8 Инверс матрицы
inverse_A = np.linalg.inv(A)
print("Обратная матрица A:\n", inverse_A)
```

```python
#9 Среднее значение
mean_A = np.mean(A)
print("Среднее значение элементов матрицы A:", mean_A)
```

```python
#10 Медиана
median_A = np.median(A)
print("Медиана элементов матрицы A:", median_A)
```

```python
#11 Стандартное отклонение
std_A = np.std(A)
print("Стандартное отклонение элементов матрицы A:", std_A)
```

```python
#12 Сумма строк
sum_rows = np.sum(A, axis=1)
print("Сумма по строкам матрицы A:", sum_rows)
```

```python
#13 Сумма колон
sum_cols = np.sum(A, axis=0)
print("Сумма по столбцам матрицы A:", sum_cols)
```

```python
#14 Максимум и минимум
data = np.array([1, 3, 7, 2, 5, 6, 4])
max_value = np.max(data)
min_value = np.min(data)
print("Максимальное значение:", max_value)
print("Минимальное значение:", min_value)
```

```python
#15 Производные
x = np.linspace(0, 10, 100)
y = x**2
dy_dx = np.gradient(y, x)
print("Градиент функции y = x^2:", dy_dx)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
titanic= titanic.iloc[:, [0, 1, 2, 3, 6, 7, 14]]
```

```python
titanic.describe()
```

```python
titanic[titanic.duplicated()]
```

```python
mean_age = titanic['age'].mean()
titanic['age'].fillna(mean_age, inplace=True)
titanic['age'] = titanic['age'].astype(int)
```

```python
mode_embarked = titanic['embarked'].mode()[0]
titanic['embarked'].fillna(mode_embarked, inplace=True)
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
titanic.drop_duplicates(inplace=True)
```

```python
titanic
```

```python
mean_age = titanic['age'].mean()
std_age = titanic['age'].std()
print(f"Средний возраст: {mean_age}, Стандартное отклонение: {std_age}")
```

```python
mean_fare = titanic['fare'].mean()
print(f"Средняя стоимость билета: {mean_fare}")
```

```python
mean_age_by_class = titanic.groupby('pclass')['age'].mean()
print(mean_age_by_class)
```

```python
print(titanic.isnull().sum())
```

```python
from scipy import stats

survived = titanic[titanic['survived'] == 1]['age']
not_survived = titanic[titanic['survived'] == 0]['age']

t_stat, p_value = stats.ttest_ind(survived, not_survived)
print(f"T-статистика: {t_stat}, P-значение: {p_value}")
```

```python
def categorize_age(age):
    if age < 10:
        return "child"
    elif age < 18:
        return "teenager"
    elif age < 40:
        return "young_adult"
    else:
        return "adult"


titanic['age_category'] = titanic['age'].apply(categorize_age)
```

```python
titanic=sns.load_dataset('titanic')
```

```python

plt.figure(figsize=(8, 6))
titanic['alone'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=['With Company', 'Alone'])
plt.title('Passenger Travel Status')
plt.axis('equal')
plt.show()
```

```python
alone_passengers = titanic[titanic['alone'] == True]

plt.figure(figsize=(10, 6))
sns.countplot(x='who', hue='class', data=alone_passengers, palette='Set2')
plt.title('Passengers Traveling Alone by Gender and Class')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='embark_town', hue='who', data=alone_passengers, palette='Set1')
plt.title('Passengers Traveling Alone by Gender and Embarked Town')
plt.xlabel('Embarked Town')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='alone', hue='survived', data=titanic)
plt.title('Survival Status by Travel Alone or with Company')
plt.xlabel('Travel Alone')
plt.ylabel('Count')
plt.xticks([0, 1], ['With Company', 'Alone'])
plt.legend(labels=['Died', 'Survived'])
plt.show()
```
